```python
# we also don't want the Id column
data.drop(['id'],axis = 1,inplace = True)
```

```python
# diagnosis column is object type so we can map it into integer
data['diagnosis'] = data['diagnosis'].map({'M' : 1, 'B' : 0})
```

```python
plt.figure(figsize =(20,20))
plt.title("Correlation Matrix")
sns.heatmap(data.corr(), color = "k", annot=True, cmap="YlGnBu" )
```

```python
X = data[['texture_mean','perimeter_mean','smoothness_mean','compactness_mean','symmetry_mean']]
y = data[['diagnosis']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
reg = LogisticRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
metrics.accuracy_score(y_test,y_pred)
```

```python
metrics.confusion_matrix(y_test,y_pred)
```
